```python
import matplotlib.pyplot as plt
import numpy as np

x_max = 3
x = np.linspace(0, x_max, 30)
y_exponential = np.exp(x)
y_linear= np.log(y_exponential) + 1
learning_per_unit = 1/y_exponential
y_learning_rate = np.ones(len(x))

plt.plot(x, y_exponential, label=r"Information units", color="#d62728")
plt.plot(x, y_linear, label=r"Learning required", color="#1f77b4")
plt.plot(x, y_learning_rate, label=r"Learning pace", color="#333", linestyle='--')
plt.plot(x, learning_per_unit, label=r"$\dfrac{Learning}{Unit}$", color="#2ca02c")

# plt.title("Taming the Information Growth Challenge").set_fontsize(12)
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend(loc=2, prop={'size': 10}, frameon=False)
plt.grid(True)

y_max_value = y_exponential.max()
y_ticks = [0,1,2,3,4] + [i for i in range(10, int(y_max_value)+1, 10)]
plt.yticks(y_ticks)
plt.xticks(np.arange(0,x_max+.1,1)) 

ax = plt.gca()
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.3)
ax.spines['bottom'].set_linewidth(0.3)

filename = '../../public/images/logarithmic_learning.png'
plt.draw()
plt.savefig(filename, dpi=300, bbox_inches='tight')

plt.show() # Must go after savefig() to avoid blank image
```
